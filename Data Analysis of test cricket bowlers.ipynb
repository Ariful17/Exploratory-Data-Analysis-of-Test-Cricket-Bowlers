**Which player played for the longest period of time?**

```python
# creating two new columns as start year and end year from span column
df['start_year']=df['Span'].str[0:4]
df['end_year']=df['Span'].str[5:]
df=df.drop('Span', axis=1)
# converting their data type into integer
df['start_year']=df['start_year'].astype('int')
df['end_year']=df['end_year'].astype('int')
# calculating their career span mathematically
df['years_played']=df['end_year']-df['start_year']
df=df.drop(['start_year', 'end_year'], axis=1)
df.head(10)
```

```python
# finding the max value of years played column
df.sort_values(by='years_played', ascending=False).head()
```
